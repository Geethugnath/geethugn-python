4.Fill in all the missing values present in all the columns in the dataset:

```python
print(df.isna().sum())
```

```python
import numpy as np



# Select only the numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Fill missing values with the mean for numeric columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Fill missing values with the mode for categorical columns
df = df.fillna(df.mode().iloc[0])
print(df.isna().sum())
```

5. Check and handle outliers in at least 3 columns in the dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Select the three columns of interest
columns_to_check = ['Age', 'Fare', 'SibSp']

# Visualize the data to identify outliers
for column in columns_to_check:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()

```

```python
# Handle the outliers using capping (flooring and ceiling)
for column in columns_to_check:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
```

```python
# Visualize the data after handling outliers
for column in columns_to_check:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column} (After Handling Outliers)')
    plt.show()
```
