1. Find out how many males and females participated in the test.

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\csaru\Downloads\StudentsPerformance.csv")
gender_count = df["gender"].value_counts()
print(gender_count)
```

2. What do you think about the students' parental level of education?

```python
import matplotlib.pyplot as plt
parental_level_count = df["parental level of education"].value_counts()
print(parental_level_count)
plt.bar(parental_level_count.index, parental_level_count.values)
plt.xticks(rotation=45)
plt.xlabel("Parental Level of Education")
plt.ylabel("Number of Students")
plt.title("Distribution of Parental Level of Education")
plt.show()
```

3. Who scores the most on average for math, reading and writing based on
â— Gender
â— Test preparation course

```python
# Gender 
gender_scores = df.groupby("gender").mean()[["math score", "reading score", "writing score"]]
print("Average Scores by Gender:\n", gender_scores)
```

```python
# Test preparation course
prep_scores = df.groupby("test preparation course").mean()[["math score", "reading score", "writing score"]]
print("\nAverage Scores by Test Preparation Course:\n", prep_scores)
```

4. What do you think about the scoring variation for math, reading and writing
based on
â— Gender
â— Test preparation course

```python
# Gender 
gender_std = df.groupby("gender").std()[["math score", "reading score", "writing score"]]
print("Standard Deviation of Scores by Gender:\n", gender_std)
```

```python
# Test preparation course
prep_std = df.groupby("test preparation course").std()[["math score", "reading score", "writing score"]]
print("\nStandard Deviation of Scores by Test Preparation Course:\n", prep_std)
```

5. The management needs your help to give bonus points to the top 25% of
students based on their math score, so how will you help the management
to achieve this.

```python
math_score_cutoff = df["math score"].quantile(0.75)
print("Math Score Cutoff for Top 25% of Students:", math_score_cutoff)
df["top 25% in math"] = df["math score"] >= math_score_cutoff
num_top_students = df["top 25% in math"].sum()
print("Number of Students in Top 25%:", num_top_students)
```
